```python
#import the necessary libraries
import pandas as pd
from nltk.tokenize import word_tokenize
from textblob import TextBlob
from string import punctuation
from nltk.tokenize import RegexpTokenizer
from googletrans import Translator
from num2words import num2words
from deep_translator import MyMemoryTranslator
import textblob.exceptions
```

```python
#Function for Removing of Punctuations or Special Characters
def remove_punctuation(text): 
    return text.translate(str.maketrans('', '', punctuation))
```

```python
#Function for Numbers to word convertion.
def num_to_words(text): 
    after_spliting = text.split()
    for index in range(len(after_spliting)):
        if after_spliting[index].isdigit():
            after_spliting[index] = num2words(after_spliting[index],lang='en')
    numbers_to_words = ' '.join(after_spliting)
    return numbers_to_words
```

```python
#function to tokenize the sentences.
def get_tokens(text):
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(text)
    return tokens
```

```python
data_set = pd.read_csv('English.csv')
```

```python
data_set.head()
```

```python
# Convert the values of 'English words/sentences' column to lowercase
datas = data_set['English words/sentences'].apply(str.lower)
```

```python
datas.head(5)
```

```python
#Removing Punctuations or Special Characters.
for i in range(len(datas)):
    datas[i]=remove_punctuation(datas[i])
```

```python
datas.head()
```

```python
#converting numbers to word.
for i in range(len(datas)):
    datas[i]=num_to_words(datas[i])
```

```python
#tokenizing the data 
tokens =[]
for i in datas:
    token = get_tokens(i)
    tokens.append(token)
```

```python
tokens
```

```python
#Function to convert English to French using MyMemoryTranslator 
def english_french(tokens):
    french = []
    for i in tokens:
        trans = MyMemoryTranslator(source='en',target='fr').translate(text=i)
        french.append(str(trans))
    text = ' '.join(french)
    return text
```

```python
french =[]
for i in tokens:
    result = english_french(i)
    french.append(result)
```

```python
#Function to convert English to spanish using TextBlob
def english_spanish(token):
    for word in token: 
        txt = ''
        try:
            w=TextBlob(word)
            w=w.translate(from_lang='en',to='es')
        except textblob.exceptions.TranslatorError:
            word=" "  
            txt=txt+word
        except textblob.exceptions.NotTranslated:
            txt=txt+word+" "
        else:
            txt=txt+str(w)+" "
    return txt
```

```python
spanish =[]
for i in tokens:
    result = english_spanish(i)
    spanish.append(result)
```

```python
#Function to convert English to german using TextBlob
def english_german(token):
    for word in token: 
        txt = ''
        try:
            w=TextBlob(word)
            w=w.translate(from_lang='en',to='de')
        except textblob.exceptions.TranslatorError:
            word=" "  
            txt=txt+word
        except textblob.exceptions.NotTranslated:
            txt=txt+word+" "
        else:
            txt=txt+str(w)+" "
    return txt
```

```python
german =[]
for i in tokens:
    result = english_german(i)
    german.append(result)
```
